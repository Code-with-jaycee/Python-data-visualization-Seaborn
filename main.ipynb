# All imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mtcars = pd.read_csv("mtcars.csv")
mtcars.head(2)
```

Create the summary of the dataset

```python
mtcars.info()
```

```python
mtcars.corr()
```

Identify the number of rows and columns 

```python
mtcars.shape
```

# Barplot

```python
# barplot
fig = plt.figure(dpi=600)
res = sns.barplot(x='cyl', y='drat', data=mtcars, palette="bright")
plt.show()
```

# Countplot

```python
fig = plt.figure(dpi=600)
sns.countplot(x='carb',data=mtcars, palette="tab10")
```

# Horizontal count plot

```python
fig = plt.figure(dpi=600)
sns.countplot(y="carb", data=mtcars, palette="rocket")
```

```python
fig = plt.figure(dpi=600)
sns.countplot(x='carb', hue='cyl', data=mtcars, palette='Set1')
```

# distplot

```python
fig = plt.figure(dpi=600)
sns.distplot(mtcars.mpg, bins=10, color='g')
```

# Heatmap


```python
# corr shows correlation of the data
# cbar draws the color bar
fig = plt.figure(dpi=600)
sns.heatmap(mtcars.corr(), cbar=True, linewidths=0.5)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris["species"].value_counts()
```

# Scatter plot

```python

fig = plt.figure(dpi=600)
sns.scatterplot(x='sepal_length', y = 'petal_length', data=iris, hue="species")
```

# Pairplot

```python

sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue="species", palette='Set1')
```

# Lmplot

```python
fig = plt.figure(dpi=600)
sns.lmplot(x='petal_length', y='petal_width',hue='species', data=iris, markers=["o", "*", "^"])
```

# Boxplot

```python
plt.figure(dpi=600)
sns.boxplot(x='species', y='sepal_width' , data=iris) 
```
